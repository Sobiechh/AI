```python
path = 'processed.hungarian.data'
```

```python
data = []
with open(path,'r') as file_o:
    line = file_o.readline()
    while line:
        tab = [float('NaN') if x=='?' else x  for x in (line.rstrip().split(','))]
        data.append(tab)
        line = file_o.readline()
```

```python
for line in data:
    print(line)
```

```python
def normalize(tab):
    min_ = (min(tab))
    max_ = (max(tab))
    
```

```python
matrix= []
for i in range(len(data[0])):
    curry_column = []
    for j in range(len(data)):
        curry_column.append(float(data[j][i]))
    normalize(curry_column)
    matrix.append(curry_column)
```

```python
t_matrix = list(zip(*matrix))
for x in t_matrix:
    print(list(x))
```
