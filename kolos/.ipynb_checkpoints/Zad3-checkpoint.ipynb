# Siec neuronowa

```python
import numpy as np #dzialamy na array
```

```python
#dane podane do zadania
X = np.array(([3, 4]), dtype=float)
X = X.reshape(1, 2) #potrzeba reshape bo sie tworzyl blad, nie byl widoczny jeden wiersz
y = np.array((1), dtype=float)
```

```python
class Neural_Network(object):
    def __init__(self):
        #Rozmiary warstw
        self.inputSize = 2 
        self.outputSize = 1
        self.hiddenSize = 2
        #do testowania
        #print(np.random.randn(self.inputSize, self.hiddenSize))
        self.W1 = np.array([[0.12, 0.10], [0.12, 0.08]]) #wagi
        self.W2 = np.array([[0.14], [0.15]]) #wagi

    def forward(self, X):
        self.z = np.dot(X, self.W1)
        self.z2 = self.sigmoid(self.z)
        self.z3 = np.dot(self.z2, self.W2)
        o = self.sigmoid(self.z3)
        return o

    #funkcja sigma
    def sigmoid(self, s):
        return 1 / (1 + np.exp(-s))

    #pochodna 
    def sigmoidPrime(self, s):
        return s * (1 - s)

    #algorytm wstecznej propagacji
    def back_prop(self, X, y, o):
        self.o_error = y - o
        self.o_delta = self.o_error * self.sigmoidPrime(o)
        self.z2_error = self.o_delta.dot(self.W2.T)
        self.z2_delta = self.z2_error * self.sigmoidPrime(self.z2)
        
        #wagi
        self.W1 += X.T.dot(self.z2_delta)
        self.W2 += self.z2.T.dot(self.o_delta)

    #trenowanie danych
    def train(self, X, y):
        o = self.forward(X)
        self.back_prop(X, y, o)
```

```python
NN = Neural_Network()
for i in range(1000): #1000 razy
    #czytelny output
    print(f"Dane wejsciowe:{(X)}")
    print(f"Dane wyjsciowe: {y}")
    print(f"Przewidziane dane wyjsciowe: {(NN.forward(X))}")
    print(f"Stracone: {(np.mean(np.square(y - NN.forward(X)))) }")
    NN.train(X, y)
```
