```python
import numpy as np
import neurolab as nl
import numpy.random as rand
import pylab as pl
```

```python
#Tworzymy zbiór danych
#Ustalamy środki naszych klas
center = np.array([[0.2, 0.2], [0.4, 0.4], [0.7, 0.3], [0.2, 0.5]])
rand_norm = 0.05 * rand.randn(100, 4, 2)
#dane uczące
inp = np.array([center + r for r in rand_norm])
inp.shape = (100 * 4, 2)
rand.shuffle(inp)

#Tworzymy dwa wejścia i 4 neurony
net = nl.net.newc([[0.0, 1.0],[0.0, 1.0]], 4)
#Uczymy sieć zgodnie z algortmem Winner Take All
error = net.train(inp, epochs=200, show=20)

#Wynik do wykrsu
w = net.layers[0].np['w']
pl.plot(inp[:,0], inp[:,1], '.', \
        center[:,0], center[:, 1] , 'yv', \
        w[:,0], w[:,1], 'p')
pl.legend(['Obiekty uczace', 'RZeczywisty srodek klasy',
           'Nauczon srodek klasy'])
pl.show()
```
